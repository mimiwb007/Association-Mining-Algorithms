```python
import pandas as pd
d1 = pd.read_csv(r"C:\Users\nibed\Desktop\Freelance\Frogs_MFCCs.csv")
d1.head()
```

```python
d2=d1[['Family','Genus','Species']]
d2.info()
```

```python
observations=[]
```

```python
for i in range(len(d2)):
    observations.append([str(d2.values[i,j]) for j in range(3)])
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(observations).transform(observations)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
frq_items = apriori(df, min_support=0.004, use_colnames=True)

```

```python
# Collecting the inferred rules in a dataframe 
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
rules
```

```python
rules = rules[['antecedents','consequents','support','lift','confidence']]
rules
```
